```python
from PIL import Image, ImageDraw, ImageChops
import colorsys
import random
import uuid
choose_func= "circle_rectangle"#@param ['circle_rectangle', 'circle', 'rectangle']
Number_of_Generate_Image=10 #@param {type: "number"}

def random_color():
    r = random.randint(0, 255)
    g = random.randint(0, 255)
    b = random.randint(0, 255)
    return (r, g, b)

def create_circle(image):
    width, height = image.size
    draw_circle = ImageDraw.Draw(image)
    circle_size=random.randint(0, height//2)
    color=random_color()
    x1_y1 = random.randint(0, width)
    x2_y2 = random.randint(0, height)
    draw_circle.ellipse((x1_y1, x2_y2, x1_y1+circle_size, x2_y2+circle_size), fill=color, outline=None)
    return image

def create_rectangle(image):
    width, height = image.size
    rectangle_width = random.randint(0, width//2)
    rectangle_height =random.randint(0, height//2)  
    draw_rect= ImageDraw.Draw(image)
    rectangle_x = random.randint(0, width)
    rectangle_y = random.randint(0, height)
    rectangle_shape = [
        (rectangle_x, rectangle_y),
        (rectangle_x + rectangle_width, rectangle_y + rectangle_height)]
    draw_rect.rectangle(
        rectangle_shape,
        fill=(
            random.randint(0, 255),
            random.randint(0, 255),
            random.randint(0, 255)
        )
    )
    return image
      
def circle_rectangle():
    rand_func=[1, 2]
    # max_element= 8 #@param {type: "number"}
    max_element= 8
    max_element=int(max_element)
    number_of_art= random.randint(1, max_element)
    background_color = random_color()
    image = Image.new('RGB', (1080, 1080),background_color)
    For_circle_rectangle_Function="both"#@param ['circle', 'rectangle','both']
    if For_circle_rectangle_Function=="both":
      for i in range(number_of_art):
          fun=random.choice(rand_func)
          if fun==1:
              image=create_circle(image)
          elif fun==2:
              image=create_rectangle(image)
    if For_circle_rectangle_Function=="circle":
      for i in range(number_of_art):
        image=create_circle(image)
    if For_circle_rectangle_Function=="rectangle":
      for i in range(number_of_art):
        image=create_rectangle(image)



    nft_name =str(uuid.uuid1())     
    image.save(f"/content/drive/MyDrive/NFT/circle_rectangle/{nft_name}.png")


def generate_circles():
    number_of_circle= random.randint(50, 200)
    image = Image.new('RGB', (1080, 1080))
    width, height = image.size
    draw = ImageDraw.Draw(image)
    circle_size=100
    for i in range(number_of_circle):
        color=(
            random.randint(0, 255),
            random.randint(0, 255),
            random.randint(0, 255)
            )
        x1_y1 = random.randint(0, width)
        x2_y2 = random.randint(0, height)
        draw.ellipse((x1_y1, x2_y2, x1_y1+circle_size, x2_y2+circle_size), fill=color, outline=None)
    nft_name =str(uuid.uuid1())     
    image.save(f"/content/drive/MyDrive/NFT/circle/{nft_name}.png")
            

def generate_rectangles():
  image = Image.new('RGB', (1080, 1080))
  width, height = image.size
  rectangle_width = 50
  rectangle_height = 50
  number_of_squares = random.randint(10, 550)
  draw_image = ImageDraw.Draw(image)
  for i in range(number_of_squares):
      rectangle_x = random.randint(0, width)
      rectangle_y = random.randint(0, height)

      rectangle_shape = [
          (rectangle_x, rectangle_y),
          (rectangle_x + rectangle_width, rectangle_y + rectangle_height)]
      draw_image.rectangle(
          rectangle_shape,
          fill=(
              random.randint(0, 255),
              random.randint(0, 255),
              random.randint(0, 255)
          )
      )
  nft_name =str(uuid.uuid1())     
  image.save(f"/content/drive/MyDrive/NFT/rectangle/{nft_name}.png")
                  



def random_color_hsv():
    h=random.random()
    s=1
    v=1
    float_rgb=colorsys.hsv_to_rgb(h,s,v)
    rgb=[int(float_rgb[0]*255),int(float_rgb[1]*255),int(float_rgb[2]*255)]
    color=tuple(rgb)
    return color
def cicle_soft():
    number_of_circle= random.randint(50, 200)
    image = Image.new('RGB', (1080, 1080))
    width, height = image.size
    draw = ImageDraw.Draw(image)
    circle_size=50
    for i in range(number_of_circle):
        overlay_image = Image.new('RGB', (1080, 1080))
        overlay_draw = ImageDraw.Draw(overlay_image)
        color=random_color_hsv()
        x1_y1 = random.randint(0, width)
        x2_y2 = random.randint(0, height)
        overlay_draw.ellipse((x1_y1, x2_y2, x1_y1+circle_size, x2_y2+circle_size), fill=color, outline=None)
        image=ImageChops.add(image,overlay_image)
    nft_name =str(uuid.uuid1())     
    image.save(f"/content/drive/MyDrive/NFT/circle/{nft_name}.png")
            




try:
  For_circle="soft color" #@param ['soft color', 'hard color']

  if choose_func=="circle_rectangle":
    for i in range(int(Number_of_Generate_Image)):
        circle_rectangle()
  if choose_func=="circle":
    if For_circle=='soft color':
      for i in range(int(Number_of_Generate_Image)):
          cicle_soft()
    if For_circle=='hard color':
      for i in range(int(Number_of_Generate_Image)):
          generate_rectangles()
  if choose_func=="rectangle":
    for i in range(int(Number_of_Generate_Image)):
        generate_rectangles()
except Exception as e:
  print(str(e))    
```
